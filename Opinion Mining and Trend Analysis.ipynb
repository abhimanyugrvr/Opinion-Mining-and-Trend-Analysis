# Using regular expression for the removal of name tags and the emoticons from tweets.

```python
def clean_tweet(tweet):
    return ' '.join(re.sub('(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)', ' ', tweet).split())
```

```python
finaldata['clean_tweet'] = finaldata['tweets_lower'].apply(lambda x: clean_tweet(x))
```
